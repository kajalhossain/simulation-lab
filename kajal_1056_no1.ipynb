```python
#2.Random Number Generation
import numpy as np
import matplotlib.pyplot as plt

# Original exam scores
exam_scores = np.array([
    93, 83, 69, 62, 75, 93, 73, 77, 65, 65,
    78, 90, 94, 78, 57, 76, 56, 78, 84, 92
])

# Calculate mean and standard deviation
mean_score = np.mean(exam_scores)
std_score = np.std(exam_scores)

# Generate random dataset using normal distribution
np.random.seed(0)  # for reproducibility
generated_scores = np.random.normal(loc=mean_score, scale=std_score, size=20)

# Plot histograms side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Real data histogram
axes[0].hist(exam_scores, bins=10, color='skyblue', edgecolor='black')
axes[0].set_title("Real Exam Scores")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Frequency")

# Generated data histogram
axes[1].hist(generated_scores, bins=10, color='salmon', edgecolor='black')
axes[1].set_title("Generated Exam Scores (Normal Dist.)")
axes[1].set_xlabel("Score")

plt.tight_layout()
plt.show()
```

```python
#3.Monte Carlo Simulation
import numpy as np

# Original exam scores
exam_scores = np.array([
    93, 83, 69, 62, 75, 93, 73, 77, 65, 65,
    78, 90, 94, 78, 57, 76, 56, 78, 84, 92
])

# Calculate mean and standard deviation
mean_score = np.mean(exam_scores)
std_score = np.std(exam_scores)

# Generate 100,000 samples from the normal distribution
np.random.seed(0)  # for reproducibility
samples = np.random.normal(loc=mean_score, scale=std_score, size=100000)

# Estimate the probability of score ≥ 75
probability = np.mean(samples >= 75)

print(f"Estimated probability that a student scores ≥ 75: {probability:.4f}")
```

```python
#4.Hypothesis Testing

import numpy as np
from scipy import stats

# Original exam scores
exam_scores = np.array([
    93, 83, 69, 62, 75, 93, 73, 77, 65, 65,
    78, 90, 94, 78, 57, 76, 56, 78, 84, 92
])

# Hypothesized mean
hypothesized_mean = 56

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(exam_scores, hypothesized_mean)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The average exam score is significantly different from 56.")
else:
    print("Fail to reject the null hypothesis: No significant difference from 56.")



```
